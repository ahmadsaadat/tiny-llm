```python
import numpy as np
```

### 1. Training Data

```python
training_sentences = [
    "the cat sat on the mat",
    "the dog ran to the house",
    "a big dog ran quickly",
    "a small cat sat slowly",
    "the cat and the dog sat",
    "the big house is red",
    "the small mat is blue",
    "the dog ran quickly to the house",
    "the cat sat slowly on the mat",
    "a red cat ran",
    "a blue dog sat",
    "the big dog and the small cat ran",
    "the house is big and red",
    "the mat is small and blue",
    "the cat ran quickly",
    "the dog sat slowly",
    "a small dog ran to the house",
    "a big cat sat on the mat",
    "the red house is big",
    "the blue mat is small",
]
```

### 2. Build Vocabulary

```python
vocab = sorted(
    set([word for sentence in training_sentences for word in sentence.split()])
)
vocab_size = len(vocab)

word_to_id = {word: i for i, word in enumerate(vocab)}

id_to_word = {i: word for word, i in word_to_id.items()}

print("vocab: ", vocab)
print("vocab_size: ", vocab_size)
```

### 3. Model Size Choices

```python
embedding_dimension = 8
n_heads = 2  # TODO: What is this
attention_dimension = embedding_dimension // n_heads  # TODO: What does this do?
max_sequence_length = 20  # TODO: What is this?

print("embedding_dimension: ", embedding_dimension)
print("n_heads: ", n_heads)
print("attention_dimension: ", attention_dimension)
```

### 4. Create Random Model Weights

```python
# These random numbers are the model's
# initial "brain" before training.

# represents words as vectors
token_embedding_table = np.random.randn(
    vocab_size,
    embedding_dimension,
)

# represents word position in a vector
position_embedding_table = np.random.randn(
    max_sequence_length,
    embedding_dimension,
)

# What am i looking for?
W_Q = np.random.randn(
    embedding_dimension,
    embedding_dimension,
)

# Who am I?
W_K = np.random.randn(
    embedding_dimension,
    embedding_dimension,
)

# What do I carry forward
W_V = np.random.randn(
    embedding_dimension,
    embedding_dimension,
)

# TODO: what does this do?
W_O = np.random.randn(
    embedding_dimension,
    embedding_dimension,
)

# TODO: what is an lm_head
lm_head = np.random.randn(
    embedding_dimension,
    vocab_size,
)
```

### 5. Create Training Examples

```python
training_examples: list[tuple] = []

for sentence in training_sentences:
    tokens = sentence.split()
    for i in range(1, len(tokens)):
        input_words = tokens[:i]
        target_word = tokens[i]

        training_examples.append((input_words, target_word))

print("Number of training examples:")
print(len(training_examples))
print("\n", training_examples)
```

### 6. Look At One Training Example

```python
input_words, target_word = training_examples[0]

print("input_words: ", input_words)
print("target_word: ", target_word)
```

### 7. Tokenize Input Words

```python
input_token_ids = [word_to_id[word] for word in input_words]

print("input_token_ids: ")
print(input_token_ids)
```

### 8. Tokenize Target Word

```python
target_token_id = word_to_id[target_word]

print("target_token_id: ")
print(target_token_id)
```

### 9. Get Token Embeddings

```python
x = token_embedding_table[input_token_ids]

print("input_token_ids: ", input_token_ids)
print("token embeddings shape: ", x.shape)
print(x[0])
```

### 10. Add Position Embeddings

```python
input_length = len(input_token_ids)

x = x + position_embedding_table[:input_length]

print("token_embeddings_table: \n\n", token_embedding_table[input_token_ids])
print("\nposition_embedding_table: \n\n", position_embedding_table[:input_length])
print("\ntoken_embedding + position embeddings: \n\n", x[0])
```

### 11. Create Attention i.e. Q, K, V

```python
Q = x @ W_Q
K = x @ W_K
V = x @ W_V

print("Q shape: ", Q.shape)
print("K shape: ", K.shape)
print("V shape: ", V.shape)
```

### 12. Split into multiple heads

```python
# Before: (input_length, embedding_dimension)
print("Q.shape before: ", Q.shape, "(token, features)")
# After reshape: Split the 8 features into multiple attention heads ()
print(
    "Q.shape after reshape: ",
    Q.reshape(input_length, n_heads, attention_dimension).shape,
    "(token, attention_heads, features)",
)
# After transpose: (heads, tokens, features)
print(
    "Q.shape after transpose: ",
    Q.reshape(input_length, n_heads, attention_dimension).transpose(1, 0, 2).shape,
    "(attention_heads, tokens, features)",
)

Q = Q.reshape(input_length, n_heads, attention_dimension).transpose(1, 0, 2)
K = K.reshape(input_length, n_heads, attention_dimension).transpose(1, 0, 2)
V = V.reshape(input_length, n_heads, attention_dimension).transpose(1, 0, 2)

print("\nQ multi-head shape:", Q.shape)
print("K multi-head shape:", K.shape)
print("V multi-head shape:", V.shape)
```

### 13. Attention Scores

```python
scores = Q @ K.transpose(0, 2, 1)

scores = scores / np.sqrt(attention_dimension)  # TODO: why do we do this here?

print("scores shape:")
print(scores.shape)
```

### 14. Create Causal Mask

```python
mask = np.triu(np.ones((input_length, input_length)) * -1e9, k=1)

scores = scores + mask

print("masked scores shape:")
print(scores.shape)
```

### 15. Attention SoftMax

```python
exp_scores = np.exp(scores - np.max(scores, axis=-1, keepdims=True))

attention_weights = exp_scores / np.sum(exp_scores, axis=-1, keepdims=True)

print("attention_weights shape:")
print(attention_weights.shape)
```

### 16. Mix Value Vectors

```python
attended = attention_weights @ V

print("attended shape: ", attended.shape)
```

### 17. Combine Heads

```python
attended = attended.transpose(1, 0, 2)

attention_output = attended.reshape(input_length, embedding_dimension)

print("attention_output shape:")
print(attention_output.shape)
```

### 18. Output Projection

```python
attention_output = attention_output @ W_O

print("projected attention_output shape:")
print(attention_output.shape)
```

### 19. Use Last Token

```python
last_token_vector = attention_output[-1]

print("last_token_vector shape:")
print(last_token_vector.shape)
```

### 20. LM Head -> Logits

```python
logits = last_token_vector @ lm_head

print("logits shape:")
print(logits.shape)

print("logits:")
print(logits)
```

### 21. Final Softmax

```python
exp_logits = np.exp(logits - np.max(logits))

probabilities = exp_logits / np.sum(exp_logits)

print("probabilities:")
print(probabilities)
```

### 22. Predicted Word

```python
predicted_token_id = int(np.argmax(probabilities))

predicted_word = id_to_word[predicted_token_id]

print("predicted_word:")
print(predicted_word)
```

### 23. Training Loss

```python
correct_probability = probabilities[target_token_id]

loss = -np.log(correct_probability)

print("target_word:")
print(target_word)

print("correct_probability:")
print(correct_probability)

print("loss:")
print(loss)
```
