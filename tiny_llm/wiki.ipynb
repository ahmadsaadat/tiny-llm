# LLM Training (Context Aware)

### 1. Load Training Data

```python
file_path = "input_data/wiki.train.raw"

with open(file_path, "r", encoding="utf-8") as f:
    raw_text = f.read()


lines = [line.strip() for line in raw_text.splitlines() if line.strip()]

for line in lines[:10]:
    print(line)
```

### 2. Word Tokenization

```python
vocab = sorted(set(word for line in lines for word in line.split()))


word_to_id = {word: idx for idx, word in enumerate(vocab)}
id_to_word = {idx: word for word, idx in word_to_id.items()}
```

### 3. Training Data Tokenization

```python
tokenized_lines = []

for line in lines:
    token_ids = [word_to_id[word] for word in line.split()]
    tokenized_lines.append(token_ids)

print("line: ", [id_to_word[word] for word in tokenized_lines[0]])
print("tokenized line: ", tokenized_lines[0])
```

### 4. Create Training Sequences

```python
# 8 previous words → next word

sequence_length = 8
training_sequences = []

for line in lines:
    words = line.split()

    token_ids = [word_to_id[word] for word in words]

    if len(token_ids) <= sequence_length:
        continue

    for i in range(len(token_ids) - sequence_length):
        input_sequence = token_ids[i : i + sequence_length]
        target_token = token_ids[i + sequence_length]

        training_sequences.append((input_sequence, target_token))

print("Number of training sequences: ", len(training_sequences))
print("\nFirst training sequence:")
print("input ids:", training_sequences[0][0])
print("target id:", training_sequences[0][1])
```
